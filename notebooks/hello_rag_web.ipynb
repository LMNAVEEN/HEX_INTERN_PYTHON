```python
from langchain_community.document_loaders import PyPDFLoader

pages = PyPDFLoader("random_20pages.pdf").load()
print(f"Loaded {len(pages)} pages")
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter

chunks = RecursiveCharacterTextSplitter(
    chunk_size=600,
    chunk_overlap=100,
    separators=["\n\n", "\n", ".", " "]
).split_documents(pages)
print(f"Created {len(chunks)} chunks")
```

```python
import chromadb
from langchain_chroma import Chroma

vector_store = Chroma.from_documents(chunks, embeddings)
print("Vector store ready")
```

```python
retriever = vector_store.as_retriever(search_kwargs={"k": 3})
```

```python
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser
from langchain_google_genai import ChatGoogleGenerativeAI
from dotenv import load_dotenv

load_dotenv()

llm = ChatGoogleGenerativeAI(model="gemini-2.0-flash-lite")

SYS_PROMPT = """
Always answer in a funny and hilarious way, always end with another relevant question.

Context:
{ctxt}
"""

prompt = ChatPromptTemplate.from_messages([
    ("system", SYS_PROMPT),
    ("human", "{question}")
])

chain = prompt | llm | StrOutputParser()

def ask_question(question_txt):
    docs = retriever.invoke(question_txt)
    context = "\n\n---\n\n".join(d.page_content for d in docs)
    return chain.invoke({"ctxt": context, "question": question_txt})

print("Ready")
```

```python
from langchain_community.document_loaders import WebBaseLoader

urls = ["https://en.wikipedia.org/wiki/Supercomputer"]
web_chunks = RecursiveCharacterTextSplitter(
    chunk_size=400,
    chunk_overlap=100
).split_documents(WebBaseLoader(urls).load())

vector_store.add_documents(web_chunks)
print(f"Added {len(web_chunks)} web chunks")
```

```python
res = ask_question("why is sea so deep?")
print(res)

res = ask_question("who invented supercomputer?")
print(res)
```
