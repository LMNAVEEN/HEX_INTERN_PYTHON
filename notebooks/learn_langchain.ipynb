# LangChain Step-by-Step Learning
## Step 1 — Plain LLM Call

**Concept:** How to talk to an LLM using LangChain.

No tools. No agents. Just you sending a message and the LLM replying.

**Flow:**
```
You → LLM → Response
```

````python
# Run this once to install dependencies
# !pip install langchain-groq python-dotenv
````

````python
from dotenv import load_dotenv
from langchain_groq import ChatGroq
from langchain_core.messages import HumanMessage, SystemMessage

# Load GROQ_API_KEY from .env file automatically
load_dotenv()

# Create the LLM
llm = ChatGroq(model="llama-3.1-8b-instant", temperature=0)

print("LLM ready!")
````

### Message Types
LangChain uses 3 message types:
| Type | Role | Purpose |
|------|------|---------|
| `SystemMessage` | system | Sets the LLM's behaviour/persona |
| `HumanMessage` | user | Your message to the LLM |
| `AIMessage` | assistant | The LLM's reply (returned automatically) |

````python
# Basic LLM call — just send a message and get a reply
response = llm.invoke([
    HumanMessage(content="What is LangChain in one sentence?")
])

print("Response:", response.content)
````

````python
# Adding a SystemMessage — gives the LLM a persona
response = llm.invoke([
    SystemMessage(content="You are a task manager assistant. Be short and helpful."),
    HumanMessage(content="I need to finish my project report by tomorrow.")
])

print("Response:", response.content)
````

````python
# Multi-turn conversation — pass the full message history
from langchain_core.messages import AIMessage

messages = [
    SystemMessage(content="You are a task manager assistant."),
    HumanMessage(content="I have 3 tasks today."),
    AIMessage(content="Sure! What are your 3 tasks?"),
    HumanMessage(content="1. Buy groceries 2. Call doctor 3. Submit report")
]

response = llm.invoke(messages)
print("Response:", response.content)
````

### What you learned in Step 1
- `ChatGroq` — connects to the Groq LLM
- `llm.invoke([...])` — sends messages and gets a response
- `SystemMessage` — sets the LLM's behaviour
- `HumanMessage` — your input
- `AIMessage` — LLM's previous reply (used in multi-turn chats)
- `response.content` — how to read the LLM's reply

**Limitation of Step 1:** The LLM can only talk — it cannot DO anything (save a task, read from DB, etc.).

➡️ **Step 2** will fix that by giving the LLM a `@tool` it can actually call.
