```python
import chromadb

client = chromadb.Client()
```

```python
collection = client.create_collection("rand_items")
```

```python
collection.add(
    ids=["id_1","id_2","id_3"],
    documents=[
        "Tom has fever",
        "Rahul is swimming",
        "Reshmi went for jogging"
    ]
)
```

```python
data = collection.get(
    include=["documents","embeddings","metadatas"]
)
```

```python
for i, emb in enumerate(data["embeddings"]):
    print(f"text {data["documents"][i]} ") 
    print(f"length {len(emb)}")
    print(f"first 10 embeddings {emb[:10]}")
```

```python
results = collection.query(
    query_texts=["who is sick today?"],
    n_results=1
)
print(results)    
```
