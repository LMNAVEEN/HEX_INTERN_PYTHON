```python
from dotenv import load_dotenv
from langchain_google_genai import ChatGoogleGenerativeAI

load_dotenv()

llm = ChatGoogleGenerativeAI(model="gemini-3.1-flash-lite")
```

```python
self_ref_prompt = """
Solve the following logic puzzle: 
"A farmer needs to cross a river with a wolf, a goat, and a cabbage. His boat can only hold himself and one item at a time. 
If left alone, the wolf eats the goat, or the goat eats the cabbage. How does he get everything across safely?"

Provide your solution, but before your final answer, include a "Self-Reflection" section where you explicitly verify:
- Did you ever leave the wolf and goat alone on a bank?
- Did you ever leave the goat and cabbage alone on a bank?
- If yes to either, rewrite your steps until the answer passes these constraints.
"""
resp = llm.invoke(self_ref_prompt)
print(resp.text)
```
